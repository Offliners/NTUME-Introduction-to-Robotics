```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from re import X
def IK(x, y):
    L1 = 10
    L2 = 8.5
    theta2 = np.arccos((x ** 2 + y ** 2 - L1 ** 2 - L2 ** 2) / (2 * L1 * L2))
    theta1 = np.arctan2(y, x) - np.arctan2(L2 * np.sin(theta2), L2 * np.cos(theta2) + L1)
    return np.degrees(theta1), np.degrees(theta2)


def TrajectoryPlanner(dx, dy, z, name):
    t_x = []
    t_y = []
    t_z = []
    d_x = []
    d_y = []
    t1 = start_point[0]
    t2 = start_point[1]
    interval = 15
    for i in range(dx.shape[0]):
        for j in range(interval):
            t1 += dx[i] * 1 / interval
            t2 += dy[i] * 1 / interval
            if t1 >= 18.5 or t2 >= 18.5:
              print('warm')
            d_x.append(t1)
            d_y.append(t2)
            temp1, temp2 = IK(t1, t2)
            t_x.append(temp1)
            t_y.append(temp2)
            t_z.append(z[i])

        if z[i]:
          isLift = 'b'
        else:
          isLift = 'dimgray'
        if i == 0:
          plt.plot(d_x[interval * i : interval * i + interval], d_y[interval * i : interval * i + interval], c = isLift)
        else:
          plt.plot(d_x[interval * i - 1: interval * i + interval], d_y[interval * i - 1 : interval * i + interval], c = isLift)

    plt.title(name)
    plt.grid()
    plt.show()

    return t_x, t_y, d_x, d_y

def plotValue(t1, t2, x, y, name):
  t = np.arange(len(t1))
  plt.figure()
  plt.subplot(411)
  plt.plot(t, t1, c='b', label=f't1 {name}')
  plt.grid()
  plt.legend()
  plt.subplot(412)
  plt.plot(t, t2, c='b', label=f't2 {name}')
  plt.grid()
  plt.legend()
  plt.subplot(413)
  plt.plot(t, x, c='b', label=f'x {name}')
  plt.grid()
  plt.legend()
  plt.subplot(414)
  plt.plot(t, y, c='b', label=f'y {name}')
  plt.grid()
  plt.legend()
  plt.show()

def plotErr(sim_x, real_x, sim_y, real_y, name1, name2):
  err_t1 = []
  err_t2 = []
  for i in range(len(sim_x)):
    err_t1.append(sim_x[i] - real_x[i])
    err_t2.append(sim_y[i] - real_y[i])

  t = np.arange(len(err_t1))
  plt.plot(t, err_t1, c = 'r', label = f'{name1} err')
  plt.plot(t, err_t2, c = 'b', label = f'{name2} err')
  plt.title('Error')
  plt.legend()
  plt.grid()
  plt.show()
  
  return np.abs(np.mean(np.array(sim_x) - np.array(real_x))), np.abs(np.mean(np.array(sim_y) - np.array(real_y)))
```

```python
sun_dx_sim = np.array([-4, 5.65685, -5.65685, -4.24264, 5.65685, 1.41421, -4.24264, 7.07107, -4.24264])
sun_dy_sim = np.array([-1, 5.65685, -5.65685, 4.24264, 5.65685, -7.07107, 4.24264, -1.41421, 4.24264])
sun_z_sim = np.array([0, 1, 0, 1, 1, 0, 1, 0, 1])

sun_dx_real = np.array([-4, 5.65685, -5.65685, -4.24264, 5.65685, 2.5, -6, 9.5, -7])
sun_dy_real = np.array([-1, 5.65685, -5.65685, 4.24264, 5.9 , -8, 6, -4.5, 7])
sun_z_real = np.array([0, 1, 0, 1, 1, 0, 1, 0, 1])

start_point = np.array([9.3, 5.2])

t1_sim, t2_sim, x_sim, y_sim  = TrajectoryPlanner(sun_dx_sim, sun_dy_sim, sun_z_sim, 'Simulate Sun')
t1_real, t2_real, x_real, y_real = TrajectoryPlanner(sun_dx_real, sun_dy_real, sun_z_real, 'Real Sun')

MAE_theta1, MAE_theta2 = plotErr(t1_sim, t1_real, t2_sim, t2_real, 'theta1', 'theta2')
MAE_x, MAE_y = plotErr(x_sim, x_real, y_sim, y_real, 'X', 'Y')
plotValue(t1_sim, t2_sim, x_sim, y_sim, 'sim')
plotValue(t1_real, t2_real, x_real, y_real, 'real')

print(f'MAE theta1 : {MAE_theta1}  MAE theta2 : {MAE_theta2}')
print(f'MAE x : {MAE_x}  MAE y : {MAE_y}')
```
